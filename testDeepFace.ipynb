```python
from deepface.basemodels import VGGFace
model = VGGFace.loadModel()
df = DeepFace.find(img_path="test.jpg", 
                   db_path="./database", 
                   enforce_detection=False, 
                   model_name = 'VGG-Face',
                   model = model,
                   distance_metric = 'cosine')
print(df)
if df.shape[0] >0:
    path = df.iloc[0].identity
    print(path)
```
